```python
def most_common(x, N):
    """outputs the N-th most common entry in a column and its count"""
    if len( x.value_counts()) > N-1:
        return x.value_counts().index[N-1]
    else: 
        return -1

def most_common_count(x, N):
    """outputs the N-th most common entry in a column and its count"""
    if len( x.value_counts()) > N-1:
        return x.value_counts().iloc[N-1]
    else: 
        return 0

def preliminary_column_analysis(df):
    """
    this function takes a and returns:
    dfcol: dataframe including stats on columns 
          (number of non NaN entries, number of distinct entries, 3 most common entries)
    """

    dfcol=df.apply(lambda x: len(x)-sum(x.isnull() ) )
    
    # number of not-Nan entries
    dfcol=pd.DataFrame(dfcol)
    dfcol.columns = ['N_not_NaN']
    
    # fraction of not_NaN entries
    dfcol['Fraction_not_NaN']=dfcol.N_not_NaN.apply(lambda x: float(x)/len(df))
    
    # unique entries
    dfcol['n_unique']=df.apply(lambda x: len(pd.Series(x).unique()))

    # most common entries
    dfcol['1st_most_common']=df.apply(lambda x: most_common(x, 1))
    dfcol['1st_most_common_count']=df.apply(lambda x: most_common_count(x, 1))
    dfcol['2nd_most_common']=df.apply(lambda x: most_common(x, 2))
    dfcol['2nd_most_common_count']=df.apply(lambda x: most_common_count(x, 2))
    dfcol['3rd_most_common']=df.apply(lambda x: most_common(x, 3))
    dfcol['3rd_most_common_count']=df.apply(lambda x: most_common_count(x, 3))

    dfcol.Fraction_not_NaN.hist(bins=20)
    plt.xlabel('Fraction of non-null')
    plt.ylabel("N columns")
    return dfcol
```
